```python
import os
import xml.etree.ElementTree as ET
from PIL import Image

# Direktori data
raw_folder = "./../data/raw/"
output_folder = "./../data/clean/"

# Loop melalui semua file XML di folder
for filename in os.listdir(raw_folder):
    if filename.endswith(".xml"):
        xml_path = os.path.join(raw_folder, filename)
        
        # Parse file XML
        tree = ET.parse(xml_path)
        root = tree.getroot()

        # Nama file gambar dan path
        img_filename = root.find("filename").text
        img_path = os.path.join(raw_folder, img_filename)
        
        # Buka gambar
        with Image.open(img_path) as img:
            # Loop untuk setiap objek yang ada di XML
            for obj in root.findall("object"):
                # Ambil nama objek (misal: "pothole")
                obj_name = obj.find("name").text
                # Ambil koordinat bounding box
                bndbox = obj.find("bndbox")
                xmin = int(bndbox.find("xmin").text)
                ymin = int(bndbox.find("ymin").text)
                xmax = int(bndbox.find("xmax").text)
                ymax = int(bndbox.find("ymax").text)
                
                # Crop gambar
                cropped_img = img.crop((xmin, ymin, xmax, ymax))
                
                # Buat folder jika belum ada
                obj_folder = os.path.join(output_folder, obj_name)
                os.makedirs(obj_folder, exist_ok=True)
                
                # Simpan hasil crop dengan nama file yang sama
                crop_filename = os.path.join(obj_folder, img_filename)
                cropped_img.save(crop_filename)
                
        print(f"Processed {img_filename} and saved crops to {obj_folder}")
```
